Processing image data for deep learning involves preparing raw images into a format suitable for training models. The steps typically include:

Image Acquisition – collecting raw image data from datasets, sensors, or web sources.

Resizing & Normalization – converting images to a fixed size and scaling pixel values (e.g., 0–1 or –1–1) for faster convergence.

Data Augmentation – applying transformations (rotation, flipping, cropping, brightness changes) to increase dataset variability and reduce overfitting.

Feature Extraction – converting images into tensors, sometimes using pretrained models (transfer learning).

Splitting Data – dividing into training, validation, and test sets for model evaluation.

These preprocessing steps ensure consistency, enhance generalization, and improve deep learning model performance in tasks like classification, detection, and segmentation.

```python
# Importing the image module from matplotlib library
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
```

Resizing the image using Pillow Library

```python
from PIL import Image
```

```python
img = Image.open('dog.jpg')
img_resized = img.resize((200, 200))
```

```python
img_resized.save('dog_resized.jpg')
```

```python
resized_image = plt.imshow(img_resized)
```

```python
img_res = mpimg.imread('dog_resized.jpg')
```

```python
img_res.shape
```

Convert RGB image to grayscale image using OpenCV

```python
import cv2 
```

```python
img = cv2.imread('dog.jpg')
```

```python
type(img)
```

```python
img.shape
```

```python
grayscale_img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
```

```python
type(grayscale_img)
```

```python
grayscale_img.shape
```

```python
# from Google.colab.patches import cv2_imshow
plt.imshow(grayscale_img)
```

```python
cv2.imwrite('dog_grayscale.jpg', grayscale_img)
```
